# Text Please

```python
from newsplease import NewsPlease
from newspaper import Article
```

```python
def text_crawl(url):
    article = NewsPlease.from_url(url)
    return article.maintext
```

```python
text_crawl(train_data['ia_link1'][19])
```

```python
train_data['text1'] = ""
err_idx = []
train_data['text2'] = ""
err_idx2 = []
```

```python
# train_data['text1'] = train_data['ia_link1'].apply(text_crawl)
for i in range(len(train_data)):
    try:
        train_data['text1'][i] = text_crawl(train_data['ia_link1'][i])
    except:
        train_data['text1'][i] = ""
        err_idx.append(i)
        print("error1: {}".format(i))
    try:
        train_data['text2'][i] = text_crawl(train_data['ia_link2'][i])
    except:
        train_data['text2'][i] = ""
        err_idx2.append(i)
        print("error2: {}".format(i))
```

```python
train_data.to_csv(index=False)
```
